```python
import pandas as pd
```

```python
ids_estation=["2117D","2331","2946X","2030","2400E","1549","2870","2465","2614"]
years=list(range(2014, 2019))
months=list(range(1,13))
```

```python
climate=pd.DataFrame()
```

```python
climate=pd.DataFrame()

for i in ids_estation:
    for year in years:
        for month in months:
            climate=pd.concat([climate, pd.read_csv("datos_clima_"+str(i)+"_"+str(year)+"_"+str(month)+".csv", index_col=0)])
```

```python
climate.index = range(climate.shape[0])
climate
```

```python
climate.to_csv("datos_clima_integrados.csv")
```
